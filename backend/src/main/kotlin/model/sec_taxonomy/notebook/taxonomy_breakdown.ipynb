```python
def extract_relationships(hierarchy):
    relationships = {}

    def traverse(concept_name, parent=None):
        if concept_name not in hierarchy:
            return

        concept_info = hierarchy[concept_name]
        sub_concepts = concept_info['sub_concepts']

        if parent:
            if parent not in relationships:
                relationships[parent] = []
            relationships[parent].append({
                'name': concept_name,
                'weight': concept_info['weight']
            })

        for sub in sub_concepts:
            traverse(sub, concept_name)

    for concept in hierarchy:
        traverse(concept)

    return relationships

relationships = extract_relationships(calculation_hierarchy)
##print(json.dumps(relationships, indent=2))
```

```python
class FinancialMetricsCalculator:
    def __init__(self, relationships):
        self.relationships = relationships

    def calculate_metric(self, metric_name, data):
        if metric_name in data:
            return data[metric_name]

        if metric_name not in self.relationships:
            raise ValueError(f"Metric {metric_name} not found in relationships.")

        related_concepts = self.relationships[metric_name]

        value = 0
        for concept in related_concepts:
            concept_name = concept['name']
            weight = float(concept['weight']) if concept['weight'] else 1
            sub_value = self.calculate_metric(concept_name, data)
            value += weight * sub_value

        return value

# Example usage
data = {
    "us-gaap:InventoryWriteDown": 100000,
    "us-gaap:InterestOnConvertibleDebtNetOfTax": 50000,
    "us-gaap:Revenues": 200000,
    "us-gaap:ConvertiblePreferredDividendsNetOfTax": 15000,
    "us-gaap:DilutiveSecurities": 5000,
    "us-gaap:DirectFinancingLeaseInterestIncome": 2000,
    "us-gaap:InterestOnConvertibleDebtNetOfTax": 500,
    "us-gaap:ConvertiblePreferredDividendsNetOfTax": 1000
}

calculator = FinancialMetricsCalculator(relationships)
total_revenue = calculator.calculate_metric("us-gaap:ProductionRelatedImpairmentsOrCharges", data)
print("Total:", total_revenue)
```
